```python
from utils.graph import Grafo
from heapq import heapify, heappop, heappush
```

```python
grafo = Grafo.load_pickled_graph("../graph/graph.pkl")
```

```python
def djisktra(grafo: Grafo, initial_node) -> dict:
    distances = distances = {node: float("inf") for node in grafo.adj_list}
    distances[initial_node] = 0

    # inicio priority queue
    pq = [(0, initial_node)]
    heapify(pq)

    visited = set()

    while pq:
        current_distance, current_node = heappop(
            pq
        )  # pega o no com a menor distancia
        if current_node in visited:
            continue
        visited.add(current_node)

        for neighbor, weight in grafo.adj_list[current_node]:
            # Calcula distancia do no atual ate o vizinho
            tentative_distance = current_distance + weight
            if tentative_distance < distances[neighbor]:
                #se achar distancia menor que atual coloca na lista de distancias
                distances[neighbor] = tentative_distance
                heappush(pq, (tentative_distance, neighbor))

    return distances


def vertices_ate_d(grafo : Grafo, node, D : float):
    distancias = djisktra(grafo, node)
    distancias_filtradas = {key: value for key, value in distancias.items() if value <= D}
    return distancias_filtradas
```

```python
distancias_ate_D = vertices_ate_d(grafo, 'jons@amerexenergy.com', 10)
print(distancias_ate_D)
```
